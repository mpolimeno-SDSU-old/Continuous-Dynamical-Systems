Problem 1) (10 pts) 3.4.11

**Solution to Problem 1**

**Part a**

We are given

$$
\dot x = rx-\sin(x) = f(x,r).
$$

For the case $r=0$

$$
\dot x = -\sin(x) = f(x).
$$

T find the fixed points we simply solve

$$
-\sin(x) = 0 \longrightarrow x_{\ast}=n\pi,
$$

where $n\in \mathbb{Z}$. Now to study their stability, take the derivative of our function

$$
f^{\prime}(x) = -\cos(x).
$$

Evaluating it at the fixed points, we see that fixed points of the form $x_{\ast}=2(n-1)\pi$ are unstable, for $n\in \mathbb{Z}$; whereas fixed points of the form $x_{\ast}=2n\pi$ are stable, for $n\in \mathbb{Z}$.
We make a phase plot to display our results.

```python
Pi=np.pi
r=0
xvals = np.linspace(-3*Pi,3*Pi,90)
yvals = np.zeros(xvals.size)
xprime = r*xvals-np.sin(xvals)
yprime = np.zeros(xvals.size)
plt.plot(xvals, r*xvals-np.sin(xvals),'--',color='b')
plt.quiver(xvals,yvals, xprime, yprime,color='r');
plt.xlabel('$x$');
```

**Part b**

We have to show that for $r>1$ we have only one fixed point.

To solve $f(x,r)=0$ we Taylor-expand $\sin(x)$ about $0$

$$
\sin(x) \approx x-\frac{x^{3}}{3!} + \mathcal{O}(x^{5}).
$$

Therfore our $f(x,r)$ becomes

$$
f(x,r) = rx-x+\frac{x^{3}}{3!}+\mathcal{O}(x^{5})
$$

Solving $f(x,r)=0$ yields

$$
x\left(r-1+\frac{x^{2}}{6}\right)=0 \longrightarrow x_{\ast}=0, x_{\ast}=\pm\sqrt{6(1-r)}.
$$

Therefore we see that our non-trivial fixed points can only exist for $r<1$. So $r_{c}=1$ is the bifurcation point. Since we are given the condition that $r>1$, then only the trivial fixed point $x_{\ast}=0$ exists. To check its stability for $r>1$ we solve

$$
f_{x}(x,r)=r-\cos(x)\longrightarrow f_{x}(0,r>1) = r-1>0,
$$

therefore $x_{\ast}=0$ is an unstable fixed point for $r>1$. We make a phase plot to display our results.

```python
Pi=np.pi
r=1.1
xvals = np.linspace(-2*Pi,2*Pi,90)
yvals = np.zeros(xvals.size)
xprime = r*xvals-np.sin(xvals)
yprime = np.zeros(xvals.size)
plt.plot(xvals, r*xvals-np.sin(xvals),'--',color='b')
plt.quiver(xvals,yvals, xprime, yprime,color='r');
plt.xlabel('$x$');
```

**Part c and Part d**

Let us check by using the only fixed point that exists for any value of $r$, i.e. $x_{\ast}=0$. We will use the result of $r_{c}=1$ from the previous part.

$$
f(x,r) = rx-\sin(x).
$$

Tangency condition

$$
f_{x} = r -\cos(x) \longrightarrow f_{x}(0,1) = 1 -\cos(0)=0.
$$

Check. Now

$$
f_{xx}(x,r) = \sin(x) \longrightarrow f_{xx}(0,1) = 0
$$

and

$$
f_{r}(x,r) = x \longrightarrow f_{r}(0,1) = 0;
$$

therefore we see that we are going to have a change in stability for the fixed point $x_{\ast}=0$ at $r_{c}=1$, which is in accord with our previous results.

We showed in Part a that, for $r<1$, $x_{\ast}=0$ in stable, whereas in Part b we showed that it was unstable for $r>1$. In fact

\begin{gather*}
f_{x}(0,r) = r - 1
\begin{cases}

we see that this resembles the normal form of the subcritical pitchfork bifurcation as the cubic term has a destabilizing effect. So we have a subcritical pitchfork bifurcation for $r_{c}=1$.

**Solution to Problem 2**


We are given

$$
\dot x = rx+x^{3}-x^{5};
$$

its potential $V(x)$ is given by


$$
-\int dV = \int (rx+x^{3}-x^{5})dx \longrightarrow V(x) =-r\frac{x^{2}}{2}-\frac{x^{4}}{4}+\frac{x^{6}}{6}
$$

To find the critical value for $r$, we need to solve $V(x)=0$. Multiplying both sides by $12$ yields

$$
-6rx^{2}-3x^{4}+2x^{6} = 0.
$$

Letting $y=x^{2}$, we can rewrite the above equation as

$$
-6ry-3y^{2}+2y^{3} = 0
$$

and solve for $y$. Thus,

$$
y(-6r-3y+2y^{2})=0,
$$

one solution is obviously $y=0$, while the other two roots are given by

$$
y_{+/-}=\frac{3\pm\sqrt{9+48r}}{4},
$$

which is real for

$$
9+48r\geq0 \longrightarrow r\geq-\frac{3}{16}.
$$

Therefore, our bifurcation value is $r_{c}=-\frac{3}{16}$, as we will not have any fixed point for any smaller value of $r$.

Problem 3) (5 pts) 3.5.3

**Solution to Problem 3**

So, we have to show that

$$
\frac{d\phi}{d\tau} = -\sin(\phi)+\gamma\sin(\phi)\cos(\phi) = f(\phi)
$$

reduces to

$$
\frac{d\phi}{d\tau} = A\phi - B\phi^{3} + \mathcal{O}(\phi^{5})
$$

near $\phi=0$.

Thus, let us Taylor-expand $\sin(\phi)$ and $\cos(\phi)$ around $\phi=0$. We get

$$
\sin(\phi) \approx \phi -\frac{\phi^{3}}{3!} + \mathcal{O}(\phi^{5})
$$

$$
\cos(\phi) \approx 1 - \frac{\phi^{2}}{2!} + \frac{\phi^{4}}{4!} + \mathcal{O}(\phi^{5})
$$

and therefore

$$
\sin(\phi)\cos(\phi) = \phi -\frac{\phi^{3}}{2!} - \frac{\phi^{3}}{3!} + \mathcal{O}(\phi^{5}).
$$

Thus,

\begin{align}
f(\phi) \approx -\phi + \frac{\phi^{3}}{3!} + \gamma\left(\phi -\frac{\phi^{3}}{2!} - \frac{\phi^{3}}{3!}\right) + \mathcal{O}(\phi^{5})\\
= -\phi + \gamma\phi - \gamma\frac{\phi^{3}}{2!} + \frac{\phi^{3}}{3!} - \gamma\frac{\phi^{3}}{3!} + \mathcal{O}(\phi^{5})\\
= (\gamma-1)\phi -\left(\frac{4\gamma-1}{3!}\right)\phi^{3} + \mathcal{O}(\phi^{5}).
\end{align}

Making $A=(\gamma-1)$ and $B=\left(\frac{4\gamma-1}{3!}\right)$ we get

$$
\frac{d\phi}{d\tau} = A\phi - B\phi^{3} + \mathcal{O}(\phi^{5}),
$$

which is the desired result.

Problem 4) (10 pts) 3.5.6

**Solution to Problem 4**

**Part a**

We have

$$
\epsilon \ddot{x} + \dot{x} + x = 0; \ \ \ x(0) = 1, \ \dot{x}(0) = 0,
$$

with $x(0)=1$ and $\dot x(0)=1$.

We use the following characteristic equation to solve the ODE

$$
\epsilon \lambda^{2} + \lambda + 1 = 0 \\
\Rightarrow \lambda_{+/-} = \frac{1}{2\epsilon}(-1 \pm \sqrt{1-4\epsilon}).
$$

Having found those $\lambda$ values, we plug them back into the original equation and get a general solution of the form

$$
x = Ae^{\lambda_{+}t} + Be^{\lambda_{-}t},
$$

where $A$ and $B$ are just constants. Using the given initial conditions, we find values for $A$ and $B$

$$
x(0) = 1 \rightarrow Ae^{\lambda_{+}(0)} + Be^{\lambda_{-}(0)} = 1 \rightarrow A + B = 1 \\
\dot{x}(0) = 0 \rightarrow A\lambda_{+}e^{\lambda_{+}(0)} + B\lambda_{-}e^{\lambda_{-}(0)} = 1 \rightarrow A\lambda_{+} + B\lambda_{-} = 1
$$

if we solve for both equations simultaneously we get 
$$
A = \frac{\lambda_{-}}{\lambda_{-}-\lambda_{+}}
$$ 

and 

$$
B = \frac{-\lambda_{+}}{\lambda_{-}-\lambda_{+}}.
$$

We then plug in those values to get the solution 

$$
x(t) = \frac{1}{\lambda_{-}-\lambda_{+}} \left(\lambda_{-}e^{\lambda_{+}t} - \lambda_{+}e^{\lambda_{-}t}\right)
$$

**Part b**

We know from literature that the presence of the parameter $\epsilon$ will yield two different time scales for our solution. To find them we Taylor-expand our $\lambda$'s. Thus,

\begin{gather*}
\lambda_{+/-} = \frac{1}{2\epsilon}(-1 \pm (1 - 2\epsilon + \mathcal{O}(\epsilon^{2}))) =
\begin{cases} 
\lambda_{+} = -1 + \mathcal{O}(\epsilon)\\
\lambda_{-} = -\frac{1}{\epsilon} + 1 + \mathcal{O}(\epsilon)
\end{cases}
\end{gather*}

Then, since we have a factor of $\frac{1}{\lambda_{-}-\lambda_{+}}$ in our solution, we solve

$$
\lambda_{-}-\lambda_{+} = -\frac{1}{\epsilon} + 1 + \mathcal{O}(\epsilon) - (-1 + \mathcal{O}(\epsilon)) = -\frac{1-2\epsilon}{\epsilon},
$$

so its reciprocal is

$$
-\frac{\epsilon}{1-2\epsilon}
$$

which we will use to make a geometric series argument at the appropriate time.

All that is left is plug these values back into our solution for $x(t)$ and do some algebra. So

$$
x(t) = -\frac{\epsilon}{1-2\epsilon}\left(\left(-\frac{1}{\epsilon} + 1 + \mathcal{O}(\epsilon)\right)e^{-t} - (-1 + \mathcal{O}(\epsilon))e^{-t/\epsilon}\right),
$$

making a geometric series argument

$$
-\frac{\epsilon}{1-2\epsilon}= -\epsilon(1-2\epsilon+\mathcal{O}(\epsilon^{2}))
$$

then

\begin{align}
x(t) = -\epsilon (1+ 2\epsilon + \mathcal{O}(\epsilon^{2}))\left(-\frac{1}{\epsilon}e^{-t}+(e^{-t}+e^{-t/\epsilon})+\mathcal{O}(\epsilon)\right)\\
=(1+ 2\epsilon + \mathcal{O}(\epsilon^{2})) \left(e^{-t} - \epsilon(e^{-t} + e^{-t/\epsilon}) + \mathcal{O}(\epsilon^{2})\right) \\
= e^{-t} + \epsilon(e^{-t} - e^{-t/\epsilon}) + \mathcal{O}(\epsilon^{2}).
\end{align}

We see that we have two different time scales in our problem: one dependent on the perturbation $\epsilon$, while the other independent from it. For $0<\epsilon<<1$, we would have $t/\epsilon$ that goes towards infinity and thus $e^{-t/\epsilon}$ would vanish for positive values of $t$ that are much greater than $\epsilon$. This seems to indicate that the long term behavior of the solution of this system should be very similar to the solution of its singular limit for $\epsilon\rightarrow0$.

The two different time scales imply that we have two different rates of decay for the solution: very fast for the portion of the solution that depends on $\epsilon$ and than slower for the portion that does not.

**Part d**

As as physical analog for a mechanical system, let us look at

$$
mL^{2}\ddot{\theta}+b\dot\theta+mgL\sin(\theta) = \tau
$$

which describes the damping-driven pendulum, where $m$ is the mass, $L$ is the length of the string at which the mass is attached to, $g$ is the acceleration due to gravity and $\theta$ is the angle with respect to the vertical. Taylor-expanding $\sin(\theta)$ and truncating the non-linearity yields

$$
\sin(\theta)\approx \theta +\mathcal{O}(\theta^{2}),
$$

which is a good approximation for small angles.

Thus, we can rewrite our ODE as


$$
mL^{2}\ddot{\theta}+b\dot\theta+mgL\theta = \tau,
$$

so that we have a linear ODE and an immediate resemblance with $\epsilon \ddot{x} + \dot{x} + x = 0$.

Scaling by $mgL$ yields

$$
\frac{L}{g}\ddot{\theta}+\frac{b}{mgL}\dot\theta+\theta = \tilde{\tau},
$$

where $\tilde{\tau}=\frac{\tau}{mgL}$.

Then, we define a dimensionless time as

$$
\tilde{t}=\frac{t}{T},
$$

then

$$
\frac{d\theta}{d\tilde{t}}\frac{d\tilde{t}}{dt} = \frac{1}{T}\frac{d\theta}{d\tilde{t}},
$$

and, similarly,

$$
\ddot{\theta}=\frac{1}{T^{2}}\frac{d^{2}\theta}{d\tilde{t}^{2}}.
$$

So, we can rewrite our ODE as

$$
\frac{L}{gT^{2}}\ddot\theta+\frac{b}{mgLT}\dot\theta+\theta=\tilde{\tau}.
$$

Then, we make the choice that the damping dominates, therefore

$$
\frac{b}{mgLT} = 1 \longrightarrow T = \frac{b}{mgL};
$$

then we define $\epsilon$ as

$$
\frac{L}{gT^{2}}=\frac{L}{g}\left(\frac{mgL}{b}\right)^{2} = \epsilon << 1.
$$

Thus, we have a linear ODE

$$
\epsilon\ddot\theta+\dot\theta+\theta=\tilde{\tau}.
$$

For the case $\tilde{\tau}=0$, we have our desired result.

Now we provide a physical analog involving an electrical circuit. Let us consider a circuit with a supplier (a battery) of electromotive force $E$, a resistor $R$, an inductor $I$ and a capacitor $C$, in series. 
The electromotive force (which is not a force, but still), is defined as

$$
L\ddot{Q}+R\dot Q +\frac{1}{C}Q=E
$$

Scaling by $C$ yeilds

$$
CL\ddot{Q}+CR\dot Q +Q={\tilde{E}},
$$

where ${\tilde{E}}=EC$.

Then, let us define a dimensionless time as

$$
\tau=\frac{t}{T},
$$

and, making the appropriate algebra as we did in the previous example, we can rewrite our ODE as

$$
\frac{CL}{T^{2}}\ddot{Q}+\frac{CR}{T}\dot Q + Q = {\tilde{E}}.
$$

Now we have a choice to make, either

$$
\frac{CL}{T^{2}}= 1,
$$

or

$$
\frac{CR}{T}=1.
$$

If we let, probably quite unrealistically, the resistance dominates over the inductive reactance, then

$$
\frac{CR}{T}=1 \longrightarrow T=CR.
$$

Thus, we can define our $\epsilon$ as

$$
\frac{CL}{T^{2}}= \frac{CL}{C^{2}R^{2}}=\frac{L}{CR^{2}} = \epsilon << 1,
$$

and our ODE becomes

$$
\epsilon\ddot{Q}+\dot Q + Q = {\tilde{E}}.
$$

For ${\tilde{E}}=0$ we get our desired result.

Following Strogatz's suggestion, we use a graphical approach to analyze the solutions of

$$
h=T\tanh^{-1}(m) -Jnm,
$$

**Part b**

We have

$$
h=T\tanh^{-1}(m) -Jnm = f(m).
$$

For the case $h=0$, we can solve for $T$ in order to find the critical temperature $T_{c}$ at which a phase transition occurs.
Let us start by looking for fixed points. 

$$
T\tanh^{-1}(m) -Jnm = 0
$$

Obviously, we have the trivial fixed point $m_{\ast}=0$, which is useful as it gives us something to Taylor-expand around and thus gives us some local information. 

Thus, Taylor-expanding $\tanh^{-1}(m)$ about $m_{\ast}=0$, we get

$$
\tanh^{-1}(m)\approx m + \frac{m^{3}}{3} + \mathcal{O}(m^{5});
$$

therefore, our $f(m)$ becomes

$$
T\left(m + \frac{m^{3}}{3}\right)-Jnm + \mathcal{O}(m^{5}) = 0,
$$


which resembles that normal form of the subcritical pitchfork bifurcation as the cubic term is a destabilizing factor. This would imply that the non-zero fixed points exist only below the bifurcation point, or, in this case, below the critical temperature. Let us see if that is the case.

$$
m\left(T-Jn+\frac{m^{2}}{3}+ \mathcal{O}(m^{4})\right) = 0,
$$

$$
m\left(T-Jn+\frac{m^{2}}{3}\right)=0 \longrightarrow m_{\ast}=0, ~ m_{\ast}=\pm\sqrt{3(Jn-T)}
$$

so, the nontrivial fixed point points exist only if 

$$
Jn-T>0 \longrightarrow T < Jn.
$$

This confirms that we are dealing with a subcritical pitchfork bifurcation as the nontrivial fixed points exist only below that threshold value. Thus, the critical temperature at which a phase transition occurs is

$$
T=Jn,
$$

which means that after the temperature, being lowered, hits that critical value, the material will spontaneously magnetized and thus be in a ferromagnetic state. Becuase of the symmetry of the system and the absence of an external magnetic field, as $h=0$, both the up and down spin configurations are equally likely, and so there are two possible ferromagnetic states.

We can check for stability by using a phase plot.

```python
J=1
n=2
T=J*n - 0.7

xvals = np.linspace(-0.99,0.99,90)
yvals = np.zeros(xvals.size)
xprime = T*np.arctanh(xvals)-J*n*xvals
yprime = np.zeros(xvals.size)
plt.plot(xvals, T*np.arctanh(xvals)-J*n*xvals,'--',color='b')
plt.quiver(xvals,yvals, xprime, yprime,color='r');
plt.xlabel('$m$');
```

The two non-trivial fixed points are located at roughly $m_{\ast}=-0.9$ and at $m_{\ast}=0.9$, and they are both unstable, as expected from the symmetry of the problem.
