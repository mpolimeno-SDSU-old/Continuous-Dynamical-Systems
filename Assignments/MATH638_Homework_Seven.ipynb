Problem 1) (5 pts) 7.1.6

**Solution to Problem 1**

**Part a**

For LC-circuits, Kirchoff's laws give us

$$
V_{L}+V_{C}=0,
$$

where $V_{L}$ is the voltage through the inductor while $V_{C}$ is the voltage through the capacitor. That relationship can be rewritten as

$$
L\dot I + \frac{Q}{C}=0.
$$

We are given that that $V_{32}=-V_{23}=V$ and, since the current is going clockwise, it can readily be seen that $V_{23}=V_{C}$, which implies that $V=-V_{C}$. Thus,

$$
V=-\frac{Q}{C}\rightarrow \dot V = -\frac{\dot Q}{C}
$$

and since $\dot Q=I$, we have

$$
\dot V = -\frac{I}{C}
$$

which is the desired result.

Also, for the total voltage $V_{tot}$ across the circut it holds

$$
L\dot I +\frac{Q}{C}+f(I) = 0
$$

and since $V=-\frac{Q}{C}$, then

$$
L\dot I -V + f(I) = 0
$$

which yields

$$
V = L\dot I + f(I)
$$

which again is the desired result.

**Part b**

We are given

$$
\tau=(LC)^{1/2}
$$

and therefore

$$
d\tau=(LC)^{-1/2}dt \rightarrow \frac{dt}{d\tau}=(LC)^{1/2}.
$$

We have derived

$$
\dot V =-\frac{I}{C}=\frac{dV}{dt}
$$

which, in terms of $\tau$ becomes

$$
\frac{dV}{d\tau}=-\frac{I}{C}(LC)^{1/2}=-\frac{I}{C^{1/2}}(L)^{1/2}.
$$

We are also given that

$$
w=VC^{1/2}
$$

and

$$
x=IL^{1/2}
$$

therefore

$$
\frac{dV}{d\tau}C^{1/2}=-IL^{1/2}
$$
$$
\Rightarrow \frac{dw}{d\tau}=-x
$$

which is the desired result.

Moreover, from $x=IL^{1/2}$ we see that $I=xL^{-1/2}$ so we can re-write

$$
V =L\dot I+f(I)
$$

as

$$
WC^{-1/2}=L\dot{I}+f(xL^{-1/2}) \longrightarrow WC^{-1/2}=L\dot{I}+F(x).
$$

We have

$$
\tau=(LC)^{-1/2}t\longrightarrow d\tau=(LC)^{-1/2}dt\longrightarrow dt = d\tau(LC)^{1/2}
$$

and so

\begin{align}
wC^{-1/2}=L\frac{dI}{dt}+F(x)
= L\frac{dI}{d\tau}(LC)^{-1/2}+F(x)
\end{align}

which yields

$$
w=L\frac{dI}{d\tau}L^{-1/2}+C^{-1/2}F(x)
$$

from 

$$
I=XL^{-1/2}
$$

we have

$$
dI=dxL^{-1/2}
$$

and thus

$$
w = L\frac{dx}{d\tau}(L^{-1/2})(L^{-1/2})+C^{-1/2}F(x) \longrightarrow w = \frac{dx}{d\tau}+C^{-1/2}F(x)
$$

Finally, letting $C^{-1/2}=\mu$ and rearranging yields

$$
\frac{dx}{d\tau}=w-\mu F(x)
$$

which is the desired result.

Problem 2) (5 pts) 7.3.9

**Solution to Problem 2**

**Part a**


We are studying the system

\begin{cases}
\begin{align}
\dot{r} = r(1-r^{2})+\mu r\cos(\theta), \\
\dot{\theta} = 1
\end{align}
\end{cases}

Using the chain rule with find the derivative of the radius $r$ with respect to $\theta$ to be

$$
\frac{dr}{dt}\cdot \frac{dt}{d\theta} = \frac{dr}{d\theta} = r(1-r^{2})+\mu r\cos(\theta).
$$

Then, using the following regular perturbation expansion to approximate the value of $r$

$$
r(\theta) = 1 + \mu r_{1}(\theta) + \mathcal{O}(\mu^{2}).
$$

we plug this new value of $r$ into

$$
\dot{r} = r(1-r^{2})+\mu r\cos(\theta),
$$

to derive


$$
\frac{d}{d\theta}(1 + \mu r_{1}(\theta) + \mathcal{O}(\mu^{2})) = (1 + \mu r_{1}(\theta) + \mathcal{O}(\mu^{2}))(1-(1 + \mu r_{1}(\theta) + \mathcal{O}(\mu^{2}))^{2}) - \mu (1 + \mu r_{1}(\theta) + \mathcal{O}(\mu^{2}))\cos(\theta) \\
\approx \mu\dot{r_{1}} - (1+\mu r_{1})(1 - (1 + \mu r_{1})^{2})- \mu (1+\mu r_{1})\cos(\theta)
$$

when the algebra settles, making sure to disregard any $\mathcal{O}(\mu^{2}) term$, we have

$$
\mu\dot{r_{1}} = - 2\mu r_{1} + \mu \cos(\theta)
$$

which can be devided by $\mu$ and yields

$$
\dot{r_{1}} = - 2 r_{1} + \cos(\theta)
$$

This first-order ODE can readily be solved by the methods of integrating factors. We have $p(t)=2$, so our integrating factor is


$$
G(\theta) = e^{\int_{0}^{\theta} 2 du} = e^{2 \theta}.
$$

therefore our solution is of the form

$$
r_{1}(\theta) = r_{10}(\theta)e^{-2\theta}+e^{-2\theta}\int_{0}^{\theta}e^{2s}{\cos(s)}ds
$$

Integrating $\int_{0}^{\theta}e^{2s}{\cos(s)}ds$ by parts yields

$$
\int_{0}^{\theta}e^{2s}{\cos(s)}ds = \frac{1}{5}(2e^{2\theta}\cos(\theta)-2+e^{2\theta}\sin(\theta))
$$

So, for $r_{1}$ we have

$$
r_{1}(\theta) = r_{10}e^{-2\theta}+\frac{1}{5}(\sin(\theta) + 2\cos(\theta) - 2e^{-2\theta})
$$

Solving for the initial condition $r_{1}(0) = r_{0}$ yields

$$
r_{1}(\theta) = r_{0}e^{-2\theta}+\frac{1}{5}(\sin(\theta) + 2\cos(\theta) - 2e^{-2\theta})
$$

So, our approximated $r$ is

$$
r = 1 + \mu\left(r_{0}e^{-2\theta}+\frac{1}{5}(\sin(\theta) + 2\cos(\theta) - 2e^{-2\theta})\right) +\mathcal{O}(\mu^{2})
$$

We can plot our results.

```python
def f(r,t,mu): 
    return r*(1-r**2) + mu*r*np.cos(t)

def rk4(f,rj,tj,dt,mu):
    k1 = f(rj,tj,mu)
    k2 = f(rj+dt*0.5*k1,tj+dt*0.5,mu)
    k3 = f(rj+dt*0.5*k2,tj+dt*0.5,mu)
    k4 = f(rj+dt*k3,tj+dt,mu)
    return rj + dt/6*(k1+(2.*k2)+(2.*k3)+k4)

def a_sol(t,r0,mu):
    r1 = (1./5.)*(np.sin(t) + 2*np.cos(t)) + (r0 - 2./5.)*np.exp(-2*t)
    return 1 + mu*r1

def ode_solver(f,r0,tf,dt,mu):
    # This program solves the ode dx/dt = f(x,t) for x(0) = x0
    # for 0<= t <= tf with time step dt.  
    
    nstep = int(np.round(tf/dt))
    sol = np.zeros(nstep)
    tvals = np.zeros(nstep)
    sol[0] = r0
    for jj in xrange(1,nstep):
        tj = (jj-1)*dt
        sol[jj] = rk4(f,sol[jj-1],tj,dt,mu)
        tvals[jj] = tj+dt
    asol = a_sol(tvals,r0,mu)
    plt.plot(tvals,sol,color='b',ls='-', label = 'Numerical')
    plt.plot(tvals,asol, color='r',ls='--', label = 'Analytical')
    plt.xlabel('$t$')
    plt.ylabel('$x(t)$')
    plt.legend(loc=(.7,.7))
```

```python
ode_solver(f,1,20,1e-3,0.01)
```

```python
ode_solver(f,.5,20,1e-3,0.5)
```

```python
ode_solver(f,.5,20,1e-3,0.9)
```

We see that for $\mu<<1$ we basically have an overlapping of the analytical and numerical solution: while there is a big discrepancy at $t=0$, the numerics matches the long term behavior of the analytical solution. Increasing the value of $\mu$ towards $1$, we see that the two solutions start to diverge: although the overall behavior is still well-described by our numerics, the error becomes significant.

Problem 3) (5 pts) 7.5.6

**Solution to Problem 3**

**Part a**

We are given the system

$$
\ddot{x}+\mu(x^{2}-1)\dot x + x = a,
$$

with $\mu>0$ and $a\in\mathbb{R}$.

Letting $\dot x = y$ we get the system

\begin{align}
\begin{cases}
\dot x = y\\
\dot y = -\mu(x^{2}-1)y-x+a
\end{cases}
\end{align}

We can readily see that the we have infinetely many fixed points of the form

$$
(x_{\ast},y_{\ast})=(a,0)
$$

since $a$ can be either positive, negative of zero. So, since the origin is a fixed point, let us start our analysis from there. The Jacobian for our system is

\begin{gather*}
J =
\begin{bmatrix}
0 & 1\\
-2\mu xy-1 & -\mu(x^{2}-1)
\end{bmatrix}
\end{gather*}

which at the origin gives us

\begin{gather*}
J =
\begin{bmatrix}
0 & 1\\
-1 & \mu
\end{bmatrix}
\end{gather*}

thus we readily see that our trace is $\tau=\mu>0$ and the determinant is $\Delta=1>0$. So now are in the first quadrant of the trace-determinant plane. Therefore we need to fuss about the sign of

$$
\tau^{2}-4\Delta = \mu^{2}-4
$$

which we see is greater than $0$ for $\mu>2$ and less than $0$ for $\mu<2$ (as negative values of $\mu$ are note allowed so we are not concerned with the negative root). So, for $\mu>2$ the origin is an unstable node, while for $\mu<2$ the origin is a stable spiral. We have therefore a change in stability at $\mu=2$ for which $\tau^{2}-4\Delta$ equals $0$ (degenerate case) and therefore $\mu_{c}=2$ is our bifurcation value.

Now, for $a\neq0$ our Jacobian is

\begin{gather*}
J =
\begin{bmatrix}
0 & 1\\
-1 & -\mu(a^{2}-1)
\end{bmatrix}
\end{gather*}

so the determinant is unchanged and still equal to $1$, but for our trace now we have


$$
\tau=-\mu(a^{2}-1);
$$

to study its sign, we noticed that, since $\mu$ is always greater than $0$, than $-\mu$ is always less than $0$; so, we can limit our analysis to the sign of $(a^{2}-1)$. Thus, we see that $\tau>0$ for $(a^{2}-1)<0$, which holds for
$-1<a<1$. So, we would have an unstable fixed point $(a,0)$ in that case. Whereas, for $a>1$ and $a<-1$ the trace would be negative and so we would be in the fourth quadrant of the trace-determinant plane, and thus we are in stable territory.
The trace would be $0$ for $|a|=1$.

**Parts b and c**

To do our numerics, we make the change of variables

$$
w = \dot x +\mu F(x)-a,
$$

where $F(x)=\frac{1}{3}x^{3}-x$. Thus we have $\dot w = \ddot{x}+\mu\dot x(x^{2}-1)=-x$ and we can write a system of ODE's as follows

\begin{align}
\begin{cases}
\dot x = w -\mu F(x)\\
\dot w = a- x
\end{cases}
\end{align}

Now, frankly, another change of variables would help the analysis and so we let $y=\frac{w}{\mu}$ and so we can re-write our system as

\begin{align}
\begin{cases}
\dot x = \mu y -\mu F(x)\\
\dot y =\frac{1}{\mu}(a-x)
\end{cases}
\end{align}

and thus we can readily find our nullclines to be $x=a$ and $y=F(x)$, the latter being our cubic nullcline.

We solve the system using fourth-order Runge-Kutta and we plot our results

```python
#a = .1
mu = 5
b = 1/3.

def f(x,t):
    F = b*(x[0])**3 - x[0]
    x1 = mu*x[1] - mu*F
    x2 = (.1-x[0])/mu
    sol = np.array([x1,x2])
    return sol

def rk4(f,xj,tj,dt):
    k1 = f(xj,tj)
    k2 = f(xj+dt*k1/2,tj+dt/2)
    k3 = f(xj+dt*k2/2,tj+dt/2)
    k4 = f(xj+dt*k3,tj+dt)
    return xj + (dt/6)*(k1+2*k2+2*k3+k4)

def ode_solver(f,x0,derx0,tf,dt):
    # This program solves the ode dx/dt = f(x,t) for x(0) = x0
    # for 0<= t <= tf with time step dt.  
    
    nstep = int(np.round(tf/dt))
    sol = np.zeros((2,nstep))
    tvals = np.zeros(nstep)
    sol[:,0] = np.array([x0, derx0])
    for jj in xrange(1,nstep):
        tj = (jj-1)*dt
        sol[:,jj] = rk4(f,sol[:,jj-1],tj,dt)
        tvals[jj] = tj+dt
    return sol
```

```python
x1 = np.linspace(-3,3,15)
x2 = np.linspace(-3,3,15)


X1,X2 = np.meshgrid(x1,x2)

t=0

u,v = np.zeros(X1.shape), np.zeros(X2.shape)

I, J = X1.shape

for ii in range(I):
    for jj in range(J):
        x = X1[ii, jj]
        y = X2[ii, jj]
        Xprime = f([x, y], t)
        u[ii,jj] = Xprime[0]
        v[ii,jj] = Xprime[1]
        
Q = plt.quiver(X1, X2, u, v, color='g')

xvals = np.linspace(-5,5,500)
mu = 5
a = .9
Fx = b*(xvals)**3 - xvals
ynullc = np.full((500),a)
xnullc = Fx

sol1 = ode_solver(f,0,0,20,1e-4)
sol2 = ode_solver(f,0,1.5,20,1e-4)
sol3 = ode_solver(f,0,-1.5,20,1e-4)

plt.plot(ynullc,xvals,ls='--',color='b',label = 'y nullcline')
plt.plot(xvals,xnullc,ls='--',color='r',label = 'Cubic nullcline')
plt.plot(sol1[0,:],sol1[1,:],color='k',ls='-',label = 'Trajectories')
plt.plot(sol2[0,:],sol2[1,:],color='k',ls='-')
plt.plot(sol3[0,:],sol3[1,:],color='k',ls='-')
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.legend(loc=(.05,.75))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

We can see that the trajectories fro either side of the cubic nullcline point towards it and we know that at the intersection between the nullcline there is a fixed point. From our previous analysis we have determined that for $\mu>2$, as in our case, and $a<1$, the fixed point will be unstable. This condtion points us towards the presence of limit cycles. In fact, we can see that trajectiories that travel towards the cubic nullcline will move along on it and then gets "captured" by the limit cycle.

Problem 4) (5 pts) 7.6.8

**Solution to Problem 4**

Following the method of two timings, we know that for equations of the form

$$
\ddot{x}+\epsilon h(x,\dot x)+x=0
$$

we can write the average equations as

$$
r^{\prime}=<h\sin(\theta)>
$$
$$
r\theta^{\prime}=<h\cos(\theta)>
$$

In our case, we are given $h(x,\dot x)=(|x|-1)\dot x$, which can be written in polar coordinates as

\begin{align}
h=(r|\cos(\theta)|-1)(-r\sin(\theta))\\
= (r\sin(\theta) - r^2|\cos(\theta)|\sin(\theta))
\end{align}

Using $r^{\prime}=\frac{1}{2\pi}\int_{0}^{2\pi}h(\theta)\sin(\theta)d\theta=<h\sin(\theta)>$ we derive

\begin{align}
r^{\prime} = < h\sin(\theta)> \\
= r<\sin^2(\theta)> - r^2<|\cos(\theta)| \sin^2(\theta)> \\
= \frac{1}{2}r - r^2\frac{1}{2\pi}\left( \int_0^{\pi/2} \cos(\theta) \sin^2(\theta) d\theta - \int_{\pi/2}^{3\pi/2} \cos(\theta) \sin^2(\theta) d\theta + \int_{3\pi/2}^{2\pi} \cos(\theta) \sin^2(\theta) d\theta\right) \\
= \frac{1}{2}r - r^2\frac{2}{3\pi}.
\end{align}

and 
\begin{align}
r\theta^{\prime} = <h\cos(\theta)> \\
= r<\sin(\theta) \cos(\theta)> - r^2<|\cos(\theta)|\cos(\theta) \sin(\theta)> \\
= 0.
\end{align}

So, we have effectively derived a system 

\begin{align}
\begin{cases}
r^{\prime}=\frac{1}{2}r - r^2\frac{2}{3\pi}\\
\theta^{\prime}=0
\end{cases}
\end{align}

So $\theta$ is a constant and plugging in the initial condition $\dot x=r\sin(\theta)=0$, we have $\theta=0$. Therefore we need to solve $r^{\prime}=\frac{1}{2}r - r^2\frac{2}{3\pi}$. We do it numerically and we get

$$ 
r(T)= \frac{3\pi e^{3\pi C1 + (1/2)T}}{4 e^{3\pi C1 + (1/2)T}+1}.
$$

Plugging in initial data $ x(0) = a $, we get $r(0) = a$ and thus the equation for $r$ simplifies to:

$$ 
r(T) = \frac{3\pi a e^{(1/2)T}}{4a(e^{(1/2)T}-1) + 3\pi}
$$

Since $T=t\epsilon$, for large values of $t$ the $1$ in the denominator is negligible and after rearrengment we have


$$ 
r(T) = \frac{3\pi  e^{(1/2)T}}{e^{(1/2)T}\left(4 + \frac{3\pi}{e^{(1/2)T}}\right)},
$$

which, for large values of $t$ simplifies to

$$
r(T) = \frac{3\pi}{4}
$$

which is are steady-state amplitude.

To solve for $x(t, \epsilon)$, we use the first order approximation:

$$ x(t,\epsilon) \approx r(T)\cos(t)$$

and therefore

$$ x(t,\epsilon) \approx \frac{3\pi a e^{0.5\epsilon t}}{4a(e^{0.5\epsilon t}-1) + 3\pi}\cos(t) $$

```python
def rk4_x(f,xj,yj,tj,dt,mu,a):
    k1 = f(xj,yj,tj,mu,a)

    k2 = f(xj+dt/2.*k1,yj,tj+dt/2,mu,a)

    k3 = f(xj+dt*k2/2.,yj,tj+dt/2,mu,a)

    k4 = f(xj+dt*k3,yj,tj+dt,mu,a)

    return xj + dt/6*(k1+2.*k2+2.*k3+k4)


def rk4_y(ff,xjj,yjj,tjj,dtt,mu,a):

    k1 = ff(xjj,yjj,tjj,mu,a)

    k2 = ff(xjj,yjj+dtt/2.*k1,tjj+dtt/2,mu,a)

    k3 = ff(xjj,yjj+dtt*k2/2,tjj+dtt/2,mu,a)

    k4 = ff(xjj,yjj+dtt*k3,tjj+dtt,mu,a)

    return yjj + dtt/6*(k1+2.*k2+2.*k3+k4)

def phaseplot3(f1,f2,x0,y0,tf,dt,mu,a): 
     
    nstep = int(np.round(tf/dt))
    
    x_sol = np.zeros(nstep)
    y_sol = np.zeros(nstep)
    tvals = np.zeros(nstep)
    
    x_sol[0] = x0
    y_sol[0] = y0
    for jj in range(1,nstep):
        # increasing time step each loop
        tj = (jj-1)*dt
        # making a sol vector and a time vector for plotting
        x_sol[jj] = rk4_x(f1,x_sol[jj-1],y_sol[jj-1],tj,dt,mu,a) # solve for x
        y_sol[jj] = rk4_y(f2,x_sol[jj-1],y_sol[jj-1],tj,dt,mu,a) # solve for y
        tvals[jj] = tj+dt
    
    return x_sol,y_sol,tvals
```

```python
def derivx(x,y,t, eps, a):
    return y # x_Dot

def derivy(x,y,t,eps, a):
    return -x - eps*(np.abs(x)-1)*y # y_dot

x_0 = 1 
y_0 = 0
eps = 0.1
for i in range(3):
    eps= .1 + .5*i
    output = phaseplot3(derivx,derivy,x_0,y_0,50,.01,eps,1) # a=1

    xvals = output[0]
    yvals = output[1]
    tvals = output[2]

    a=1
    x_approx = (3*np.pi*np.exp(.5*eps*tvals))/(4*a*(np.exp(.5*eps*tvals)-1)+3*np.pi)*np.cos(tvals)
    
    plt.figure(i)
    plt.title('$\epsilon$ = %f'%eps )
    plt.plot(tvals,xvals, color = 'red', label= 'numerical')
    plt.plot(tvals,x_approx, ls = '--', color ='blue', label = 'approximate')
    plt.ylim(-10,10)
    plt.xlim(0,50)
    plt.legend()
```

So we see that our numerical solution gives a good approximation of our analytical solution even for $\epsilon$ as big as $1.1$, even though they start diverge at $t\approx25$, but the matches for smaller values of $\epsilon$ and the small divergence for larger values of $\epsilon$ is a good indicator of the accuracy of our results.

**Solution to Problem 5**

**Part a**

We are given

$$
\ddot{x}+\epsilon\dot x^{3}+x=0
$$

which means we have

$$
h(x,\dot x) = \dot x^{3}=-r^{3}\sin^{3}(\theta),
$$

where for the last equality we have used polar coordinates.

Thus, for the averaged equations we have

\begin{align}
r^{\prime} = <h\sin(\theta)>\\
= <-r^{3}\sin^{4}(\theta)>\\
=-\frac{3}{8}r^{3}
\end{align}

and

\begin{align}
r\phi^{\prime}=<h\cos(\theta)>\\
= <-r^{3}\sin^{3}(\theta)\cos(\theta)>\\
= 0
\end{align}

**Part b**

We are given $x(0)=a$ and $\dot x(0)=0$. So

$$
r(0)\approx \sqrt{x(0)^{2}+\dot x(0)^{2}} = a
$$

and

$$
\frac{dr}{d\tau}=r^{\prime},
$$

where $\tau=\epsilon t$.

Then, following our results from Part a

$$
\frac{dr}{d\tau}=-\frac{3}{8}r^{3}
$$

which can readily be solved by separating the variables and yields

$$
r(t) = \frac{1}{\sqrt{\frac{3}{4}\epsilon t + \frac{1}{r^{2}(0)}}},
$$

where $r(0)$ is the amplitude of oscillation at $t=0$. Now we go back to Cartesian coordinates to get

$$
x(t) = \frac{\cos(t)}{\sqrt{\frac{3}{4}\epsilon t + \frac{1}{a^{2}(0)}}},
$$

And we can finally plot our results both analytically and numerically for $\epsilon=2$ and $a=1$

**Part c**

```python
a = 1
ep = 2

def f(x,t):
    x1 = x[1]
    x2 = -ep*x[1]**3 - x[0]
    sol = np.array([x1,x2])
    return sol

def rk4(f,xj,tj,dt):
    k1 = f(xj,tj)
    k2 = f(xj+dt*k1/2,tj+dt/2)
    k3 = f(xj+dt*k2/2,tj+dt/2)
    k4 = f(xj+dt*k3,tj+dt)
    return xj + (dt/6)*(k1+2*k2+2*k3+k4)

def an_x(t):
    x = np.sqrt(1./(.75*ep*t + (1/a**2.)))*np.cos(t)
    return x

def ode_solver(f,x0,derx0,tf,dt):
    # This program solves the ode dx/dt = f(x,t) for x(0) = x0
    # for 0<= t <= tf with time step dt.  
    
    nstep = int(np.round(tf/dt))
    sol = np.zeros((2,nstep))
    tvals = np.zeros(nstep)
    sol[:,0] = np.array([x0, derx0])
    for jj in xrange(1,nstep):
        tj = (jj-1)*dt
        sol[:,jj] = rk4(f,sol[:,jj-1],tj,dt)
        tvals[jj] = tj+dt
    asol = an_x(tvals)
    plt.plot(tvals,sol[0,:],color='r',ls='--',label='Numerics')
    plt.plot(tvals,asol,color='b',ls='-',label='Analytical Solution')
    plt.xlabel('$t$')
    plt.ylabel('$x(t,\epsilon)$')
    plt.legend(loc=(0.4,0.7))
```
