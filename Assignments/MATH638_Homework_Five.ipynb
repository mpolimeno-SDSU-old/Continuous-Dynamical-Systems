Problem 1) (5 pts) 4.3.3

**Solution to Problem 1**

We have

$$
\dot\theta = \mu\sin(\theta)-\sin(2\theta);
$$

first, let us look for fixed points as they provide some pieces of information on the dynamics:

$$
\mu\sin(\theta)-2\sin(\theta)\cos(\theta)=0
$$

which yields to

$$
\sin(\theta)=0\longrightarrow \theta_{\ast}=0, ~ n\pi
$$

where $n\in \mathbb{Z}$. And

$$
\mu = 2\cos(\theta) \longrightarrow \theta_{\ast}=\cos^{-1}\left(\frac{\mu}{2}\right),
$$

which is bounded above and below by the values of $\mu$. Therefore we see that all the possible values of $\mu$ for bifurcation to occur are

$$
-2\leq\mu\leq 2,
$$

Now, letting $\dot\theta=f(\theta,\mu)$, we compute the stability of those fixed points.

$$
f_{\theta}=\mu\cos(\theta)-2\cos(2\theta)
$$

for $\theta=0,n\pi$ with $n$ even, we have

$$
\mu-2=0
$$

which means that the bifurcation value for those fixed points is $\mu_{c}=2$. The existence of those fixed points is independent of $\mu$, therefore we see that they are stable for $\mu<2$ and unstable for $\mu>2$.

For $n$ odd we have

$$
-\mu - 2 = 0 \longrightarrow \mu_{c}=-2
$$

where $\mu_{c}$ is the bifurcation point for those fixed points. They are unstable for $\mu<-2$ and stable for $\mu>-2$.

Now, for the fixed points of the form

$$
\theta_{\ast}=\cos^{-1}\left(\frac{\mu}{2}\right)
$$

we have

$$
f_{\theta}\left(\cos^{-1}\left(\frac{\mu}{2}\right)\right) = \mu\cos\left(\cos^{-1}\left(\frac{\mu}{2}\right)\right) - 2\cos\left(2\left(\cos^{-1}\left(\frac{\mu}{2}\right)\right)\right) = 0.
$$

and, after some trigonometric identities

\begin{align}
\frac{\mu^{2}}{2} - 2\left(\cos^{2}\left(\cos^{-1}\left(\frac{\mu}{2}\right)\right) - \sin^{2}\left(\cos^{-1}\left(\frac{\mu}{2}\right)\right)\right) = 0\\
\frac{\mu^{2}}{2} - 2\left(\frac{\mu^{2}}{4} - \left(1 - \frac{\mu^{2}}{4}\right)\right) = 0\\
\frac{\mu^{2}}{2} - 2\left(\frac{\mu^{2}}{2} - 1\right) = 0 \\
-\frac{\mu^{2}}{2} + 2 = 0.
\end{align}

So, as anticipated in our analysis above, these fixed points exists only for $-2<\mu< 2$. The bifurcation occurs are the upper and lower bound for the $\mu$ values, and thus $\mu_{c}=\pm2$.

Moreover, Taylor-expanding our original ODE around the trivial fixed point $\theta_{\ast}=0$ yields

\begin{align}
\mu\left(\theta-\frac{\theta^{3}}{3!}+\mathcal{O}(\theta^{5})\right)-2\left(\theta-\frac{\theta^{3}}{2}-\frac{\theta^{3}}{3!}+\mathcal{O}(\theta^{5})\right)\\
=(\mu-2)\theta-\left(\mu+\frac{4}{3}\right)\theta^{3}+\mathcal{O}(\theta^{5})
\end{align}

which is the normal form of a subcritical pitchfork bifurcation. This information added to the fact that the non trivial fixed points exist only below $\mu=2$ leads to the conclusion that we are in fact dealing with a subcritical pitchfork bifurcation.

We make several phase plots to display the various changes in stability of the fixed points as $\mu$ varies.

Problem 2) (5 pts) 4.3.10

**Solution to Problem 2**

We are given

$$
\dot x = r + rx^{2n},
$$

where $n>1$ and $\in \mathbb{Z}$, and we have to show that

$$
T_{bottleneck} \approx cr^{b},
$$

and find $b,~ c$.

So, from the lecture we know that

$$
T \tilde{}\mathcal{O}(r^{-1/2});
$$

thus, let

$$
x=r^{a}u,
$$

where $u\tilde{}\mathcal{O}(1)$ and $x$ has a characteristic scale $\mathcal{O}(r^{a})$, with $a$ unknown. Then, suppose

$$
t=r^{b}\tau,
$$

where $\tau\tilde{}\mathcal{O}(1)$.

Then, it follows

$$
dx=r^{a}du, ~ dt=r^{b}d\tau \longrightarrow \frac{dx}{dt}=\frac{r^{a}}{r^{b}}\frac{du}{dt}=r^{a-b}\frac{du}{d\tau}
$$

So, our $\dot x$ becomes

$$
\dot x = r^{a-b}\frac{du}{d\tau} \longrightarrow \dot x = r + r^{2an}u^{2n}.
$$

Let every term be of the same order with respect to $r$, then

\begin{align}
r\tilde{} \mathcal{O}(1)\\
r^{2an}\tilde{} \mathcal{O}(1)\\
r^{a-b}\tilde{} \mathcal{O}(1)\\
\end{align}

therefore


$$
2an=1 \longrightarrow a=\frac{1}{2n}
$$

and

$$
a-b=1 \longrightarrow \frac{1}{2n}-b = 1 \longrightarrow b=\frac{1}{2n}-1,
$$

which is our value for $b$.

Now, recall that our $\dot x$ in terms of $u$ and $\tau$ is

$$
r+ru^{2n}=r\frac{du}{d\tau}
$$

which implies

$$
(1+u^{2n})=\frac{du}{d\tau} \longrightarrow d\tau =\frac{du}{1+u^{2n}}
$$

thus

$$
\tau = \int_{-\infty}^{\infty}\frac{du}{1+u^{2n}}
$$

Since,

$$
t=r^{b}\tau
$$

letting $\tau=c$ will give us the desired result. And, therefore, our $T_{bottleneck}$ is

$$
T_{bottleneck}\approx r^{\frac{1}{2n}-1}\int_{-\infty}^{\infty}\frac{du}{1+u^{2n}}.
$$

Problem 3) (5 pts) 4.4.1

**Solution to Problem 3**

The non-linear ODE given is

$$
mL^{2}\ddot{\theta}+b\dot\theta+mgL\sin(\theta)=\Gamma,
$$

where $\Gamma$ is the torque, whose units are a force ($mg$) times a length ($L$).

Therefore, let us rescale the ODE by a factor of $mgL$ to get

$$
\frac{L}{g}\ddot{\theta}+\frac{b}{mgL}\dot\theta+\sin(\theta)=\tilde{\Gamma},
$$

where $\tilde{\Gamma}=\frac{\Gamma}{mgL}$.

Then, let us introduce a non-dimensional time

$$
t=\frac{\tilde{t}}{T},
$$

so that we can rewrite our ODE as

$$
\frac{L}{gT^{2}}\ddot{\theta}+\frac{b}{mgLT}\dot\theta+\sin(\theta)=\tilde{\Gamma};
$$

we now make the choice that, in our system, damping dominates over gravity, therefore

$$
\frac{b}{mgLT}=1 \longrightarrow \frac{b}{mgL}=T,
$$

and thus

$$
\frac{L}{gT^{2}}\longrightarrow\frac{L}{g}\left(\frac{mgL}{b}\right)^{2}\longrightarrow\frac{m^{2}gL^{3}}{b^{2}}.
$$

Let us call

$$
\frac{m^{2}gL^{3}}{b^{2}}=\epsilon
$$

so that we can rewrite our ODE as

$$
\epsilon\ddot{\theta}+\dot\theta+\sin(\theta)=\tilde{\Gamma},
$$

if we make $0<\epsilon<<1$, which implies that damping has a much bigger impact on the system than gravity does, then the $\ddot{\theta}$ term is negligible and we can approximate our ODE as

$$
\dot\theta+\sin(\theta)=\tilde{\Gamma},
$$

which is the overdamped limit for the system under study.

Problem 4) (5 pts) 5.2.2

**Solution to Problem 4**


**Part a**


We have the system

\begin{align}
\dot x=x-y\\
\dot y=x+y
\end{align}

Therefore, the matrix $A$ for the system is

\begin{gather*}
A =
\begin{bmatrix}
1 & -1\\
1 & 1 
\end{bmatrix}
\end{gather*}

The trace is given by

$$
\tau = 1+1=2
$$

and the determinant is

$$
\Delta=(1*1)-(-1*1)=2
$$

therefore, since

$$
\tau^{2}-4\Delta = 4-8=-4<0
$$

we are dealing with complex eigenvalues.

To find them, we just need to solve the characteristic polynomial

$$
p(\lambda)=\lambda^{2}-\tau\lambda+\Delta=0 \longrightarrow p(\lambda) = \lambda^{2}-2\lambda+2=0
$$

which leads to

$$
\lambda_{+}=1+i, ~ \lambda_{-}=1-i.
$$

To find the eigenvectors, we need to solve

$$
A{\bf x}=\lambda{\bf x}
$$

for both eigenvalues.

$\lambda=1+i$ leads to

\begin{gather*}
{\bf v_{1}} =
\begin{bmatrix}
i\\
1
\end{bmatrix},
\end{gather*}

whereas $\lambda=1-i$ yields

\begin{gather*}
{\bf v_{2}} =
\begin{bmatrix}
-i\\
1
\end{bmatrix},
\end{gather*}

**Part b**

The solution for the system is of the form

$$
{\bf x}(t)= c_{1}e^{\lambda_{+} t}{\bf v_{1}}+c_{2}e^{\lambda_{-}t}{\bf v_{1}},
$$

which is in terms of complex coefficients. To express our solution in terms or real-valued functions, we notice that our ${\bf x}(t)$ is a linear combination of two lineary independent solutions, therefore we restrict our analysis to 

$$
{\bf x_{1}}(t) = e^{\lambda_{+} t}{\bf v_{1}} \longrightarrow {\bf x_{1}}(t) = e^{(1+i) t}\begin{bmatrix}
i\\
1
\end{bmatrix},
$$

which is still a solution for the system. Thus, using Euler's formula we get

\begin{align}
{\bf x_{1}}(t) = e^{t}(\cos(t)+i\sin(t))\begin{bmatrix}i\\
1
\end{bmatrix}\\
= e^{t}\begin{bmatrix}
i\cos(t)-\sin(t)\\
\cos(t)+i\sin(t)
\end{bmatrix}.
\end{align}

Separating reals and imaginary parts, and putting aside the $e^{t}$ term to simplify the notation, for now, yields


\begin{gather*}
Re({\bf x_{1}}(t))=
\begin{bmatrix}
-\sin(t)\\
\cos(t)
\end{bmatrix},
\end{gather*}

\begin{gather*}
Im({\bf x_{1}}(t))=
\begin{bmatrix}
\cos(t)\\
\sin(t)
\end{bmatrix}
\end{gather*}

which are both real-valued eigenvectors. Therefore any linear combination of them will yield a solution to the system. Thus

$$
{\bf x}(t) = c_{1}e^{t}\begin{bmatrix}
-\sin(t)\\
\cos(t)
\end{bmatrix}
+ c_{2}e^{t}\begin{bmatrix}
\cos(t)\\
\sin(t)
\end{bmatrix}
$$

which is the desired result.

Problem 5) (5 pts) 5.2.11

**Solution to Problem**

We have to show that the matrix

\begin{gather*}
A=
\begin{bmatrix}
\lambda & b\\
0 & \lambda
\end{bmatrix}
\end{gather*}

has only one-dimensional eigenspace corresponding to eigenvalue $\lambda$.

Then, since $\lambda$ is an eigenvalue, then it needs to hold

$$
A{\bf x}=\lambda{\bf x},
$$

where

\begin{gather*}
{\bf x}=
\begin{bmatrix}
x\\
y
\end{bmatrix}.
\end{gather*}

Thus, matrix multiplication yields

\begin{align}
\lambda x + by = \lambda x\\
\lambda y = \lambda y
\end{align}

therefore, our only eigenvector is

\begin{gather*}
{\bf v} =
\begin{bmatrix}
1\\
0
\end{bmatrix}.
\end{gather*}

The solution is given by

$$
{\bf x}(t) = c_{1}e^{\lambda t}{\bf v}
$$

```python
l = -2 #lambda value chosen
b = 3
def f(X,t):
    x1,x2 = X
    return [l*x1+b*x2, l*x2]

x1 = np.linspace(-2,2,15)
x2 = np.linspace(-2.5,2.5,15)

X1,X2 = np.meshgrid(x1,x2)

t=0

u,v = np.zeros(X1.shape), np.zeros(X2.shape)

I, J = X1.shape

for ii in range(I):
    for jj in range(J):
        x = X1[ii, jj]
        y = X2[ii, jj]
        Xprime = f([x, y], t)
        u[ii,jj] = Xprime[0]
        v[ii,jj] = Xprime[1]
        
Q = plt.quiver(X1, X2, u, v, color='r') #plot the vector field

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim([-2.5,2.5]);
```

Problem 6)(5 pts) 5.2.13

**Solution to Problem 5**

Our second order ODE is

$$
m\ddot{x} + b\dot x + kx = 0.
$$

Let $\dot x = v$ and we can rewrite the second-order ODE as a system of two first-order ODEs

\begin{align}
\dot x = v\\
\dot v = -\frac{k}{m}x-\frac{b}{m}v
\end{align}

Which yields the matrix

\begin{gather*}
A =
\begin{bmatrix}
0 & 1\\
-\frac{k}{m} & -\frac{b}{m}
\end{bmatrix},
\end{gather*}

for which we have trace and determinant, respectively, equal to

$$
\tau=-\frac{b}{m},
$$

$$
\Delta=\frac{k}{m}.
$$

And eigenvalues given by

$$
\lambda_{+/-}=\frac{1}{2}\left(\tau\pm\sqrt{\tau^{2}-4\Delta}\right)
$$

Thus, all, or at least most, of what we need to know to describe the dynamics of the system will be given by the sign of $\tau^{2}-4\Delta$

The following vector fields analyze all the cases that might arise and plots several solutions. The code provide comments to illustrate the different cases.

```python
k=50  #determinant = delta = k/m > 0 #trace = tau = -b/m < 0 
m=10  #tau^2 -4delta = 5 > 0 # real eigenvalues
b=50  #damping constant positive, thus tau negative #no negative mass allowed
      #stable node at the origin
    
def f(X,t): #define system of first order ODE
    x1,x2 = X
    return [x2,-(k/m)*x1 - (b/m)*x2]

x1 = np.linspace(-2,2,50)
x2 = np.linspace(-2.5,2.5,50)

X1,X2 = np.meshgrid(x1,x2)

t=0

u,w = np.zeros(X1.shape), np.zeros(X2.shape)

I, J = X1.shape

for ii in range(I):
    for jj in range(J):
        x = X1[ii, jj]
        v = X2[ii, jj]
        vprime = f([x, v], t)
        u[ii,jj] = vprime[0]
        w[ii,jj] = vprime[1]
        
Q = plt.quiver(X1, X2, u, w, color='r') #plot the vector field

plt.xlabel('$x$')
plt.ylabel('$v$')
plt.ylim([-2.5,2.5])

from scipy.integrate import odeint #plot some solutions
l = [-2.0,-1.5,-1.0,-0.5,0,0.5, 1, 1.5, 2.0]
for x20 in l:
    for x10 in l:
        tspan = np.linspace(0, 20, 600) #make them look as smooth as possible #600 is good
        x0 = [x10, x20]
        xs = odeint(f, x0, tspan)
        plt.plot(xs[:,0], xs[:,1], 'b-') # path
        plt.plot([xs[-1,0]], [xs[-1,1]], 'o',color='k') # end #attractive origin
    

plt.xlim([-2, 2])

plt.show()
```
