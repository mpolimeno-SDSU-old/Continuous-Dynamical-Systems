Problem 2) (5 pts) 6.3.11
**Answer** 
We have the system:

$$
\dot{r} = -r \\
\dot{\theta} = \frac{1}{\ln|r|}
$$

with initial conditions:

$$
r(0) = r_{0} \\
\theta(0) = \theta_{0}.
$$

**a&b)** 

$$
\frac{dr}{dt} = -r \Rightarrow \frac{1}{-r}dr = dt
$$

which we then use 

$$
-\int\frac{1}{-r}dr = \int dt \\
\Rightarrow -\ln|r| = t + c.
$$

Using the initial condition $r(0) = r_{0}$, we get the solution 

$$
r(t) = r_{0}e^{-t}.
$$

We now use this to solve for 

$$
\frac{d\theta}{dt} = \frac{1}{\ln\left|r_{0}e^{-t}\right|} \\
\Rightarrow \frac{d\theta}{dt} = \frac{1}{\ln\left|r_{0}\right| - t} \\
\Rightarrow d\theta = \frac{dt}{\ln\left|r_{0}\right| - t} \\
\Rightarrow \int d\theta = \int \frac{dt}{\ln\left|r_{0}\right| - t} \\
\Rightarrow \theta = -\ln\left|\ln(r_{0}) - t\right| + c
$$

If use the initial condition, $\theta(0) = \theta_{0}$, we get the value for c and solution:

$$
\theta = -\ln\left|\ln(r_{0}) - t\right| + \theta_{0} + \ln\left|\ln\left(\frac{1}{r_{0}}\right)\right|.
$$

Looking at these functions, if we take the limit $t \rightarrow \infty$, $r(t) \rightarrow 0$ and $\theta(t) \rightarrow \infty$.

**c)** To convert to cartesian coordinates we use 

$$
x = r\cos(\theta); ~ ~ ~ y = r\sin(\theta)
$$

which if we derive with the chain rule we get 

$$
\dot{x} = \dot{r}\cos(\theta) - r \dot{\theta}\sin(\theta), \\
\dot{y} = \dot{r}\sin(\theta) + r \dot{\theta}\cos(\theta)
$$

which gives us

$$
\dot{x} = -r\cos(\theta) - r \frac{1}{\ln|r|}\sin(\theta), \\
\dot{y} = -r\sin(\theta) + r \frac{1}{\ln|r|}\cos(\theta)
$$

which if we use $r = \sqrt{x^{2}+y^{2}}$, we get 

$$
\dot{x} = -\sqrt{x^{2}+y^{2}}\cos(\theta) - \sqrt{x^{2}+y^{2}} \frac{1}{\ln|\sqrt{x^{2}+y^{2}}|}\sin(\theta), \\
\dot{y} = -\sqrt{x^{2}+y^{2}}\sin(\theta) + \sqrt{x^{2}+y^{2}} \frac{1}{\ln|\sqrt{x^{2}+y^{2}}|}\cos(\theta).
$$

Which using the relation for x,y with respect to theta from above, we get

$$
\dot{x} = -x + \frac{y}{\ln|\sqrt{x^{2}+y^{2}}|} \\
\dot{y} = -y - \frac{x}{\ln|\sqrt{x^{2}+y^{2}}|}.
$$

If we linearize this system, we get

$$
\dot{x} = -x \\
\dot{y} = -y.
$$


Problem 3) (5 pts) 6.5.7

**Answer** We have the equation

$$
\frac{d^{2}u}{d\theta^{2}} + u = \alpha + \epsilon u^{2}.
$$

**a&b&c&d)** Which we then rewrite as a 2D system

$$
\dot{u} = v, \\
\dot{v} = \alpha + \epsilon u^{2} - u.
$$

The equilibrium points of the system are found by solving the system 

$$
v = 0, \\
\alpha + \epsilon u^{2} - u = 0
$$

which gives us the equilibrium points $(v_{*},u_{*}) = \left(0, \frac{1 \pm \sqrt{1 - 4\epsilon \alpha}}{2\epsilon}\right)$.

If we linearize using the jacobian we get 

$$
J = \left(\begin{matrix}
0 & 1\\
2\epsilon \alpha - 1 & 0\\
\end{matrix}\right)
$$

which we then evaluate at the equilibirum points 

$$
J\left(0, \frac{1 \pm \sqrt{1 - 4\epsilon \alpha}}{2\epsilon}\right) = \left(\begin{matrix}
0 & 1\\
\pm \sqrt{1 - 4\epsilon \alpha} & 0\\
\end{matrix}\right)
$$

If we find the eigenvalues for the jacobian at $\left(0, \frac{1 + \sqrt{1 - 4\epsilon \alpha}}{2\epsilon}\right)$, we get $\lambda_{+} = \pm (1 - 4\epsilon \alpha)^{1/4}$ which gives us a saddle at that equilibrium as long as $\alpha > \frac{1}{4\epsilon}$. If we evaluate the eigenvalues for the jacobian at $\left(0, \frac{1 - \sqrt{1 - 4\epsilon \alpha}}{2\epsilon}\right)$, we get $\lambda_{-} = \pm \sqrt{-\sqrt{1 - 4\epsilon \alpha}}$ which will give us imaginary eigenvalues with no real part, so we have a center. The fact that the 2D system is reversible gives us a nonlinear center at that point. 

Problem 4) (5 pts) 6.5.15 

**Answer** 
**a)**
We have 

$$
mr\ddot{\phi} = -b\dot{\phi} - mg\sin(\phi) + mr\omega^{2}\sin(\phi)\cos(\phi)
$$

and since $b = 0$ we get the equation

$$
mr\ddot{\phi} = - mg\sin(\phi) + mr\omega^{2}\sin(\phi)\cos(\phi).
$$

If we nondimensionalize with respect to $\tau = t\omega$ we get

$$
mr\omega^{2}\phi^{''} = - mg\sin(\phi) + mr\omega^{2}\sin(\phi)\cos(\phi) \\
\Rightarrow \phi^{''} = \frac{-g}{r\omega^{2}}\sin(\phi) + \sin(\phi)\cos(\phi) \\
\Rightarrow \phi^{''} = \sin(\phi)(\cos(\phi) - \gamma^{-1})
$$

**b)** phase portraits shown below

**c)** We can see that after $\gamma > 1$ we start getting what appears to be a supercritical pitchfork bifurcation for the fixed point (0,0). When $\gamma < 1$ we have the resting point as our fixed point so the bead will rest at the bottom, but once $\gamma$ crosses that threshold the bead will stop at angles given by the pitchfork bifurcation. 

```python
gamma = .7
theta = np.linspace(-4,4,50)
yvals = np.zeros(theta.size)
xprime = np.sin(theta)*(np.cos(theta) - (1./gamma))
yprime = np.zeros(theta.size)
plt.plot(theta,xprime,ls='--',color='k')
plt.quiver(theta,yvals,xprime,yprime,color='r')
plt.xlabel('$\\theta$')
plt.ylabel('$\dot{\\theta}$')
plt.grid()
plt.show()
```

Problem 5) (5 pts) 6.7.4

**Answer** We have 

$$
\ddot{\theta} + \sin\theta = 0
$$

with an amplitude of $\alpha$.

**a)** We use the trick

$$
\dot{\theta}\ddot{\theta} + \dot{\theta}\sin\theta = 0 \\
\Rightarrow \frac{d}{dt}\left(\frac{1}{2}\dot{\theta}^{2} - \cos\theta \right) = 0 \\
\Rightarrow \frac{1}{2}\dot{\theta}^{2} - \cos\theta + \cos\alpha = 0\\
\Rightarrow \dot{\theta}^{2} = 2(\cos\theta - \cos\alpha) \\
\Rightarrow \frac{d\theta}{dt} = \left(2(\cos\theta - \cos\alpha)\right)^{1/2}\\
\Rightarrow T = 4\int_{0}^{\alpha}\frac{d\theta}{\left(2(\cos\theta - \cos\alpha)\right)^{1/2}}
$$

**b)** we then use the half-angle identity:

$$
\sin\left(\frac{\alpha}{2}\right) = \sqrt{\frac{1-\cos\alpha}{2}} \Rightarrow -2\sin^{2}\left(\frac{\alpha}{2}\right) + 1 = \cos\alpha
$$

to rewrite the integral as

$$
T = 4\int_{0}^{\alpha}\frac{d\theta}{\left(2(-2\sin^{2}\left(\frac{\theta}{2}\right) + 2\sin^{2}\left(\frac{\alpha}{2}\right))\right)^{1/2}} \\
= 4\int_{0}^{\alpha}\frac{d\theta}{\left(4(\sin^{2}\left(\frac{\alpha}{2}\right) - \sin^{2}\left(\frac{\theta}{2}\right))\right)^{1/2}}
$$

using the substitution, 

$$
(\sin\left(\frac{\alpha}{2}\right))\sin\phi = \sin\left(\frac{\theta}{2}\right) \\
\Rightarrow (\sin\left(\frac{\alpha}{2}\right))\cos\phi d\phi = \frac{1}{2}\cos\left(\frac{\theta}{2}\right)d\theta \\
d\theta = \frac{2(\sin\left(\frac{\alpha}{2}\right))\cos\phi}{\cos\left(\frac{\theta}{2}\right)} d\phi
$$

which gives us 

$$
T = 4\int_{0}^{\frac{\pi}{2}}\frac{\frac{2(\sin\left(\frac{\alpha}{2}\right))\cos\phi}{\cos\left(\frac{\theta}{2}\right)} d\phi}{\left(4(\sin^{2}\left(\frac{\alpha}{2}\right) - \sin^{2}\left(\frac{\alpha}{2}\right)\sin^{2}\left(\frac{\phi}{2}\right))\right)^{1/2}} \\
= 4\int_{0}^{\frac{\pi}{2}}\frac{\frac{2(\sin\left(\frac{\alpha}{2}\right))\cos\phi}{\cos\left(\frac{\theta}{2}\right)} d\phi}{2\sin\left(\frac{\alpha}{2}\right)\cos\phi} \\
= 4\int_{0}^{\frac{\pi}{2}}\frac{d\phi}{\cos\left(\frac{\theta}{2}\right)} = 4\int_{0}^{\frac{\pi}{2}}\frac{d\phi}{\sqrt{1 - \sin^{2}\left(\frac{\theta}{2}\right)}} \\
$$

which gives us the elliptical integral of the first order

$$
T = 4\int_{0}^{\frac{\pi}{2}}\frac{d\phi}{\sqrt{1 - \sin^{2}\left(\frac{\alpha}{2}\right)\sin^{2}\left(\phi\right)}}
$$ 

where we let $m = \sin^{2}\left(\frac{\alpha}{2}\right)$.

So, $T = 4k(m)$ where

$$
k(m) = 4\int_{0}^{\frac{\pi}{2}}\frac{d\phi}{\sqrt{1 - m\sin^{2}\left(\phi\right)}}; ~ ~ ~ 0 \leq m \leq 1.
$$
